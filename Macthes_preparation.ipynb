```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv(r"C:\Users\hulsu\Downloads\IPL Data\matches.csv")
df.head()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df.isnull().sum()
```

```python
df['result_margin'].fillna(0, inplace=True)
df['target_runs'].fillna(0, inplace=True)
df['target_overs'].fillna(0, inplace=True)
```

```python
df['method'].unique()
```

```python
df['method'].fillna('None',inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['winner'].isnull()]
```

```python
df['winner'].fillna('No result',inplace=True)
df['player_of_match'].fillna('None',inplace=True)
```

```python
df[df['city'].isnull()]['venue'].unique()
```

```python
df.loc[df['venue'] == 'Dubai International Cricket Stadium', 'city'] = df.loc[df['venue'] == 'Dubai International Cricket Stadium', 'city'].fillna('Dubai')
df.loc[df['venue'] == 'Sharjah Cricket Stadium', 'city'] = df.loc[df['venue'] == 'Sharjah Cricket Stadium', 'city'].fillna('Sharjah')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['city'].unique()
```

```python
df['city'] = df['city'].replace({
    'Bangalore': 'Bengaluru',
    'Chandigarh': 'Mohali'
})
```

```python
df['city'].nunique()
```

```python
df['team1'].unique()
```

```python
df['team2'].unique()
```

```python
team_map = {
    'Royal Challengers Bangalore': 'RCB',
    'Royal Challengers Bengaluru': 'RCB',
    'Kolkata Knight Riders':'KKR',
    'Mumbai Indians':'MI',
    'Chennai Super Kings':'CSK',
    'Kings XI Punjab':'KXIP',
    'Rajasthan Royals':'RR',
    'Delhi Daredevils':'DD',
    'Sunrisers Hyderabad':'SRH',
    'Deccan Chargers': 'DCH',
    'Delhi Capitals':'DC',
    'Pune Warriors':'PW',
    'Punjab Kings':'PK',
    'Rising Pune Supergiant':'RPS',
    'Rising Pune Supergiants':'RPS',
    'Gujarat Lions':'GL',
    'Kochi Tuskers Kerala':'KTK',
    'Gujarat Titans': 'GT',
    'Lucknow Super Giants':'LSG',
    'No result':'NR'
}
```

```python
df['team1'] = df['team1'].map(team_map)
df['team2'] =df['team2'].map(team_map)
df['toss_winner'] = df['toss_winner'].map(team_map)
df['winner'] = df['winner'].map(team_map)
```

```python
df['team1'].unique(),df['team2'].unique()
```

```python
df.head()
```

```python
df.columns
```

```python
df['Teams'] = df['team1'] + ' vs ' + df['team2']
```

```python
df.head()
```

```python
# Select required columns
df_selected = df[['id', 'date', 'match_type','team1','team2', 'venue', 'toss_winner', 'toss_decision', 
                  'winner', 'result', 'result_margin', 'target_runs', 'player_of_match', 'umpire1', 'umpire2']].copy()

# Create 'teams' column by combining 'team1' and 'team2'
df_selected['teams'] = df['team1'] + ' vs ' + df['team2']

# Calculate 1st innings score (target_runs - 1)
df_selected['1st_innings_score'] = df['target_runs'] - 1

# Calculate 2nd innings score (win margin + 1st innings score)
df_selected['2nd_innings_score'] =  df_selected['1st_innings_score'] - df['result_margin']

# Reorder the columns with 'player_of_match' before umpires
df_selected = df_selected[['id', 'date', 'match_type', 'teams','team1','team2', 'venue', 'toss_winner', 
                           'toss_decision', 'winner', 'result', 'result_margin', 
                           '1st_innings_score', '2nd_innings_score', 'player_of_match', 'umpire1', 'umpire2']]

# Display the cleaned dataframe
df_selected.head()
```

```python
df_selected.dtypes
```

```python
df_selected['result_margin'] = df_selected['result_margin'].astype('Int64')
df_selected['1st_innings_score'] = df_selected['1st_innings_score'].astype('Int64')
df_selected['2nd_innings_score'] = df_selected['2nd_innings_score'].astype('Int64')
```
